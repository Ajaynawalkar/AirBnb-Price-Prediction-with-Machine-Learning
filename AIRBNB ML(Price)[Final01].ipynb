### Feature Data Types
- Nominal Data Type = Nominal Data is used to label variables without any order or quantitative value. The color of hair can be considered nominal data, as one color can’t be compared with another color.
- Ordinal Data Type = Ordinal data have natural ordering where a number is present in some kind of order by their position on the scale. These data are used for observation like customer satisfaction, happiness, etc., but we can’t do any arithmetical tasks on them. 
- Discrete Data Type = The term discrete means distinct or separate. The discrete data contain the values that fall under integers or whole numbers. 
- Continues Data Type = Continuous data are in the form of fractional numbers. It can be the version of an android phone, the height of a person, the length of an object, etc. Continuous data represents information that can be divided into smaller levels. 
- Binary Data Type = The term which provide categorical information under Yes or No, 0 or 1, Pass or Fail condition.

```python
df["reviews per month"].fillna(df["reviews per month"].mean(),inplace=True)
df["minimum nights"].fillna(df["minimum nights"].mean(),inplace=True)
df["availability 365"].fillna(df["availability 365"].mean(),inplace=True)
df["calculated host listings count"].fillna(df["calculated host listings count"].mean(),
                                            inplace=True)
df["number of reviews"].fillna(df["number of reviews"].median(), inplace=True)
```

### Fill null date 

```python
num_days = 10
g = df['last review'].notna().cumsum()
days = pd.to_timedelta(df.groupby(g).cumcount().mul(num_days), unit='d')
df['last review'] = df['last review'].ffill().add(days)
```

# K-Fold Cross Validation

Cross-validation is a technique for validating the model efficiency by training it on the subset of input data and testing on previously unseen subset of the input data. We can also say that it is a technique to check how a statistical model generalizes to an independent dataset.

- Generalization refers to your model's ability to adapt properly to new, previously unseen data, drawn from the same distribution as the one used to create the model.

- Due to high accuracy of model that's why we check with k fold cross validation in it.

## Import Cross_Val_Score Using Scikit Learn Library.

```python
from sklearn.model_selection import cross_val_score
```
